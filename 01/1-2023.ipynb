The Problem is to extract numbers from strings in a text file, but we need to only take the first and last numer. we're going to ultimately use char.isdigit(), but we need to get it into a list first so that we can take the list and run it as input in a for loop.

```python
with open('calibration.txt', 'r') as file:
    lines = file.readlines()

with open('calibration.txt', 'r') as file:
    lines = [line.strip() for line in file.readlines()]

# Ok, so this just demonstrates that the lines setup works.
#print(lines)
```

Now we build the loop

```python
full_count = []

for value in lines:
    first_digit = next((char for char in value if char.isdigit()), None)
    last_digit = next((char for char in reversed(value) if char.isdigit()), None)
    if first_digit and last_digit:
        full_count.append(int(f"{first_digit}{last_digit}"))

total = sum(full_count)
print(total)

```
